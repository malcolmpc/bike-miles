```python
import pandas as pd
from scipy.spatial.distance import cdist
```

```python
def closest_point(point, points):
    """ Find closest point from a list of points. """
    return points[cdist([point], points).argmin()]

def match_value(df, col1, x, col2):
    """ Match value x from col1 row to value in col2. """
    return df[df[col1] == x][col2].values[0]


df1 = datedset
df2 = df

df1['point'] = [(x, y) for x,y in zip(df1['SITE_LONGITUDE'], df1['SITE_LATITUDE'])]
df2['point'] = [(x, y) for x,y in zip(df2['lon'], df2['lat'])]

df2['closest'] = [closest_point(x, list(df1['point'])) for x in df2['point']]
df2['AQI'] = [match_value(df1, 'point', x, 'DAILY_AQI_VALUE') for x in df2['closest']]
```
